```python
import numpy
import random 
import math
import seaborn as sns
import matplotlib.pyplot as plt

#2.generating arrays
uniform_distribution = numpy.random.uniform(0, 1, 1000) 
normal_distribution_m0_sd1 = numpy.random.normal(0, 1, 1000) 
normal_distribution_m0_sd10 = numpy.random.normal(0, 10, 1000)

#beta_distribution_a0_b0 = numpy.random.beta(0, 0, 1000)
beta_distribution_a2_b2 = numpy.random.beta(2, 2, 1000)
beta_distribution_a3_b9 = numpy.random.beta(3, 9, 1000)

mean1 = [0,0]
E1 = [[1,0],[0,1]]
E2 = [[2,1],[1,2]]
xE1, yE1 = numpy.random.multivariate_normal(mean1, E1,10000).T
xE2, yE2 = numpy.random.multivariate_normal(mean1, E2,10000).T
```

```python
#3 Compraring histograms with plots of analytical formulas
#uniform distribution
plt1 = plt.hist(uniform_distribution, density = True)
y1 = (1/(1-0))
plt11 = plt.hlines(y1, xmin=0, xmax=1, color = 'r')
plt.show()
```

```python
#normal distribution with mean = 0 and standard deviation = 1
n1, bins1, patches1 = plt.hist(normal_distribution_m0_sd1, density = True)
mean1 = 0
sd1 = 1
normal_distribution_1 = (1/(sd1 * math.sqrt(numpy.pi*2))) * numpy.exp(-0.5*((bins1-mean1)/sd1)**2)
plt2_1 = plt.plot(bins1, normal_distribution_1, color = 'r')
plt.show()
```

```python
#normal distribution with mean = 0 and standard deviation = 10
n2, bins2, patches2 = plt.hist(normal_distribution_m0_sd10, density = True)
mean2 = 0
sd2 = 10
normal_distribution_2 = (1/(sd2 * math.sqrt(numpy.pi*2))) * numpy.exp(-0.5*((bins2-mean2)/sd2)**2)
plt3_1 = plt.plot(bins2, normal_distribution_2, color = 'r')
plt.show()
```

```python
#beta distribution_1 -> alpha1 = 2 , beta1 = 2
n3, bins3, patches3 = plt.hist(beta_distribution_a2_b2, density = True)
alpha1 = 2
beta1 = 2
beta_distribution_1 = ((bins3**(alpha1-1)) * ((1-bins3)**(beta1- 1))) / (((math.factorial(alpha1-1))*(math.factorial(beta1-1)))/(math.factorial((alpha1+beta1)-1)))
plt4_1 = plt.plot(bins3, beta_distribution_1, color = 'r')
plt.show()
```

```python
#beta distribution_2 -> alpha2 = 3, beta2 = 9
n4, bins4, patches4 = plt.hist(beta_distribution_a3_b9, density = True)
alpha2 = 3
beta2 = 9
beta_distribution_2 = ((bins4**(alpha2-1)) * ((1-bins4)**(beta2 - 1))) / (((math.factorial(alpha2-1))*(math.factorial(beta2-1)))/(math.factorial((alpha2+beta2)-1)))
plt4_1 = plt.plot(bins4, beta_distribution_2, color = 'r')
plt.show()
```

```python
#4 Two dimensional distribution
#with E1
plt5 = plt.hist2d(xE1, yE1, bins = 100)
plt.show()
```

```python
#with E2
plt6 = plt.hist2d(xE2, yE2, bins = 100)
plt.show()
```

```python
#5 Generate 1000 y points....
alpha = numpy.random.normal(2, 0.1, 1000) 
beta = numpy.random.normal(0.2, 0.2, 1000) 
x = numpy.random.normal(0, 3, 1000) 
sigma = abs(numpy.random.normal(0, 1, 1000))
mi = alpha + (beta*x)
y = numpy.random.normal(mi, sigma , 1000)  
```

```python
#6 Plot values 
plt7 = plt.plot(x,y)
plt.xlim(-6,6)
plt8 = plt.plot(x,mi)
plt.xlim(-6,6)
plt.show()
```
