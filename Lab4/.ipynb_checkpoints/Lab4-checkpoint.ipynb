```python
from cmdstanpy import CmdStanModel
import arviz as az
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
_BASE_URL = "https://raw.githubusercontent.com/rmcelreath/rethinking/Experimental/data"
HOWELL_DATASET_PATH = f"{_BASE_URL}/Howell1.csv"
d = pd.read_csv(HOWELL_DATASET_PATH, sep=';', header=0)
d=d[d.age>=18] #just adults 
d.head()
```

```python
model_ppc = CmdStanModel(stan_file='height_1_ppc.stan')

R = 1000;
sim = model_ppc.sample(iter_sampling = R,
                       iter_warmup = 0, 
                       chains = 1, 
                       fixed_param = True,
                       seed = 29042020, refresh = R)
```

```python
hist_mu = plt.hist(sim.mu)
plt.xlabel("Height[cm]")
plt.ylabel("Number of people")
```

```python
hist_sigma = plt.hist(sim.sigma)
plt.xlabel("Error[cm]")
plt.ylabel("Number of people")
```

```python
hist_height = plt.hist(sim.height)
plt.xlabel("Height[cm]")
plt.ylabel("Number of people")
```

```python
sns.jointplot(x = sim.mu.squeeze(), y = sim.sigma.squeeze())
```

```python
model_1_fit = CmdStanModel(stan_file = "height_1_fit.stan")
```

```python
fit = model_1_fit.sample(data=dict(N=len(d),
                                  heights=d.height.values),
                                seed = 28052020)
```

```python
sns.jointplot(x = fit.mu.squeeze(), y = fit.sigma.squeeze())
```

```python
hist_mu = plt.hist(fit.mu)
plt.xlabel("Height[cm]")
plt.ylabel("Number of people")
```

```python
hist_sigma= plt.hist(fit.sigma)
plt.xlabel("Error[cm]")
plt.ylabel("Number of people")
```

```python
hist_heights = plt.hist(fit.height)
plt.xlabel("Height[cm]")
plt.ylabel("Number of people")
```

```python
d['c_weight'] = d['weight'] - d['weight'].mean()
data_sim={'N':50, 'weight':np.linspace(d.c_weight.min(),d.c_weight.max())}
```

```python
model_ppc=CmdStanModel(stan_file='height_2a_ppc.stan')
R = 1000
sim_1=model_ppc.sample(data=data_sim, 
                     iter_sampling=R, 
                     iter_warmup=0, 
                     chains=1, 
                     refresh=R,
                     fixed_param=True,
                     seed=29042020)
```

```python
weight = np.linspace(d.c_weight.min(),d.c_weight.max())

for i in range(len(sim_1.alpha)):
    y = sim_1.alpha[i] + sim_1.beta[i] * weight
    plt.plot(weight,y)
    
plt.hlines(0, xmin = - 14 , xmax = 18,  color = 'r')
plt.hlines(272, xmin = - 14 , xmax = 18,  color = 'r')
plt.ylabel('Height[cm]')
plt.xlabel('Weight')
```

```python
model_2_fit=CmdStanModel(stan_file='height_2_fit.stan')
N = 50;
data_fit = data_fit = {'N':N, 'heights': d['height'][:N].values, 'weight':np.linspace(d['c_weight'][:N].min(),d['c_weight'][:N].max())};
fit=model_2_fit.sample(data=data_fit,seed=28052020)
```

```python
for i in range(len(sim_1.alpha)):
    y_3 = fit.alpha[i] + fit.beta[i] * weight
    plt.plot(weight, y_3)
plt.ylabel('Height [cm]')
plt.xlabel('Weight')
```

```python
x=data_fit['weight']
heights_mean = np.mean(fit.height, axis=0)
heights_std = np.std(fit.height, axis=0)

plt.errorbar(x, heights_mean, yerr=heights_std, fmt='x', color='b', label='Simulated values')
plt.plot(x, data_fit['heights'], 'x', color='k', label='Observed values')
plt.xlabel('Centered weight')
plt.ylabel('Height[cm]')
plt.legend()
```
